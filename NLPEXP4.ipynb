```python
pattern = r'\s'
input['twitts'].apply(lambda x:re.sub(pattern,'',x))
```

```python
pattern = r'\s'
input['Char_count'] = input['twitts'].apply(lambda x:len(re.sub(pattern,'',x)))
input
```

```python
input['Word_count'] = input['twitts'].apply(lambda x:len(x.split()))
input
```

```python
input['avg_word_length'] = input['Char_count']/input['Word_count']
input
```

```python
input['avg_word_length'].apply(lambda x: round(x,2))
```

```python
input['hash_tags_count'] = input['twitts'].apply(lambda x:len(re.findall(r'#\w+',x)))
input
```

```python
input['mentions_count'] = input['twitts'].apply(lambda x:len(re.findall(r'@\w+',x)))
input
```

FOR EMAILS

```python
pattern = r'\w+@\w+.\w+'
input['twitts'].apply(lambda x:re.findall(pattern=pattern,string = x))
```

```python
pattern =r'\b[A-Za-z._+-]+@[A-Za-z.].[A-Z|a-z]{2,}\b' #word boundary
input['email']= input['twitts'].apply(lambda x:re.findall(pattern=pattern,string = x))
```

```python
pattern =r'\b[A-Za-z0-9._%+-]+@[A-ZA-ZA-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b' #word boundary
input['email']= input['twitts'].apply(lambda x:",".join(re.findall(pattern=pattern,string = x)))
input
```

```python
pattern =r'http\S+|www\S+'
input['URLs']= input['twitts'].apply(lambda x:",".join(re.findall(pattern=pattern,string = x)))
input['URL_count']= input['twitts'].apply(lambda x:x.count(',')+1 if len(x)>0 else 0)
input
```
