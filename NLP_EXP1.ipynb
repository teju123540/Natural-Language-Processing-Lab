```python
import nltk
nltk.download('punkt_tab')

```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

word tokenization

```python
input ="Hi i am Teju. How are you?"
output1 = word_tokenize(input)
print(output1)
```

sentence tokenization

```python
output2 = sentence_tokenize(input)
print(output2)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
input = "Hi i am devil"
doc = nlp(input)
```

```python
for token in doc:
  print(token.text)
```

```python
for sent in doc.sents:
  print(sent)
```
