```python
data = "Natural Language Processing (NLP) is a field of artificial intelligence that enables computers to understand, interpret, and generate human language"
```

TOKENIZATIION

```python
data = nltk.sent_tokenize(data)
data
```

LOWER CASE AND TEXT CLEANING


```python
import re
```

```python
for i in range(len(data)):
    data[i] =data[i].lower()
    data[i]=re.sub(r'\W','',data[i])
    data[i]=re.sub(r'\s+','',data[i])
```

```python
data
```

HISTOGRAM BUILDING (FINDING FREQUENT FORM OF WORDS)


```python
word2count = {}
for para in data:
  words = nltk.word_tokenize(para)
  for word in words:
    if word not in word2count.keys():
      word2count[word] = 1
    else:
      word2count[word]+= 1
```

```python
word2count
```

```python
import heapq
freq_words = heapq.nlargest(50,word2count,key=word2count.get)
```

```python
freq_words
```

```python
import nltk
nltk.download('punkt')

dataset = ['Natural Language Processing (NLP) is a field of artificial intelligence that enables computers to understand, interpret, and generate human language']

freq_words = ["language","nlp","artificial","computers","human"]

X = []

for data in dataset:
    vector = []
    for word in freq_words:
        if word in nltk.word_tokenize(data.lower()):
            vector.append(1)
        else:
            vector.append(0)
    X.append(vector)

print(X)
```
